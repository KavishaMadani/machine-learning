Hence a univariate model and polynomial with degree 2 will underfit the given data. While polynomial model with degree 3 and 4 will be a good fit for the given data. While polynomial model with degree greater than 4 will overit the given data.

```python
def f(x):
    return x + 2 * np.sin(0.5 * x)
x = np.linspace(-10, 10, 50)
noise_mean = 0
noise_variance = 0.5
noise = np.random.normal(noise_mean, np.sqrt(noise_variance), 50)
y_noise = f(x) + noise
x_test = np.linspace(-10, 10, 100)
y_test_true = f(x_test)
train_errors = []
test_errors = []
for degree in range(11):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(x.reshape(-1, 1), y_noise)
    y_train_pred = model.predict(x.reshape(-1, 1))
    train_error = mean_squared_error(y_noise, y_train_pred)
    train_errors.append(train_error)
    y_test_pred = model.predict(x_test.reshape(-1, 1))
    test_error = mean_squared_error(y_test_true, y_test_pred)
    test_errors.append(test_error)
plt.figure(figsize=(10, 6))
degrees = range(11)
plt.plot(degrees, train_errors, label='Training Error', marker='o')
plt.plot(degrees, test_errors, label='Generalization Error', marker='o')
plt.xlabel('Model Capacity (Degree of Polynomial)')
plt.ylabel('Mean Squared Error')
plt.title('Training and Generalization Error vs. Model Capacity')
plt.xticks(degrees)
plt.legend()
plt.grid(True)
plt.show()
```

```python
degrees = np.arange(11)
bias_squared = np.zeros(len(degrees))
variance = np.zeros(len(degrees))
training_error = np.zeros(len(degrees))
generalization_error = np.zeros(len(degrees))
for degree in degrees:
  y_pred_all_datasets = np.zeros((200, 100))
  for i in range(100):
    x_train = np.random.uniform(-10, 10, 50)
    y_train = f(x_train) + np.random.normal(0, np.sqrt(0.5), 50)
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(x_train.reshape(-1, 1), y_train)
    x_test = np.random.uniform(-10, 10, 200)
    y_test_true = f(x_test)
    y_pred = model.predict(x_test.reshape(-1, 1))
    y_pred_all_datasets[:, i] = y_pred
    y_pred_mean = np.mean(y_pred_all_datasets, axis=1)
    bias_squared[degree] = np.mean((y_pred - f(x_test)) ** 2)
    variance[degree] = np.mean(np.var(y_pred_all_datasets, axis=1))
    y_train_pred = model.predict(x_train.reshape(-1, 1))
    training_error[degree] = mean_squared_error(y_train, y_train_pred)
    generalization_error[degree] = mean_squared_error(y_test_true, y_pred_mean)
plt.figure(figsize=(10,6))
plt.plot(degrees, bias_squared, label='Bias^2', marker='o')
plt.plot(degrees, variance, label='Variance', marker='o')
plt.plot(degrees, training_error, label='Training Error', marker='o')
plt.plot(degrees, generalization_error, label='Generalization Error', marker='o')
plt.xlabel('Model Capacity (Degree of Polynomial)')
plt.ylabel('Error')
plt.title('Bias^2, Variance, Training Error, and Generalization Error vs. Model Capacity')
plt.xticks(degrees)
plt.legend()
plt.grid(True)
plt.show()
```

```python
x_test = np.linspace(-10, 10, 100)
y_test_true = f(x_test)
overfitting_degree = 9
x_more = np.linspace(-10, 10, 1000)
y_more = f(x_more) + np.random.normal(0, np.sqrt(0.5), 1000)
model = make_pipeline(PolynomialFeatures(overfitting_degree), LinearRegression())
model.fit(x.reshape(-1, 1), y_noise)
model_more = make_pipeline(PolynomialFeatures(overfitting_degree), LinearRegression())
model_more.fit(x_more.reshape(-1, 1), y_more)
y_test_pred = model.predict(x_test.reshape(-1, 1))
y_test_pred_more = model_more.predict(x_test.reshape(-1, 1))
plt.figure(figsize=(10, 6))
plt.scatter(x_more, y_more, label='Datapoints with More Data', alpha=0.6)
plt.plot(x_test, y_test_true, label='True Function')
plt.plot(x_test, y_test_pred, label='Original Model (Overfitting)')
plt.plot(x_test, y_test_pred_more, label='Model with More Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('True Function vs. Model Predictions (Original & More Data)')
plt.legend()
plt.grid(True)
plt.show()
print("Mean squared error (original model):", mean_squared_error(y_test_true, y_test_pred))
print("Mean squared error (model with more data):", mean_squared_error(y_test_true, y_test_pred_more))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
lambdas = np.arange(0, 5.5, 0.5)
train_errors, test_errors, num_non_zero_w = [], [], []

for lam in lambdas:
  model = Lasso(alpha=lam)
  model.fit(x_train.reshape(-1, 1), y_train)
  y_train_pred = model.predict(x_train.reshape(-1, 1))
  train_error = mean_squared_error(y_train, y_train_pred)
  train_errors.append(train_error)
  y_test_pred = model.predict(x_test.reshape(-1, 1))
  test_error = mean_squared_error(y_test, y_test_pred)
  test_errors.append(test_error)
  num_non_zero_w.append(np.count_nonzero(model.coef_))

plt.figure(figsize=(10, 6))
plt.plot(lambdas, train_errors, label='Training Error', marker='o')
plt.plot(lambdas, test_errors, label='Test Error', marker='o')
plt.xlabel('Lambda')
plt.ylabel('Mean Squared Error')
plt.title('Training and Test Error vs. Lambda')
plt.xticks(lambdas)
plt.grid(True)
plt.show()
```

```python
print("Training error (lambda=1.5):", train_errors[3])
print("Test error (lambda=1.5):", test_errors[3])
print("Number of non-zero coefficients (lambda=1.5):", num_non_zero_w[3])
```

```python
np.random.seed(0)
x = np.linspace(-10, 10, 100)
y = f(x)
epsilon = np.random.normal(0, 0.5, 100)
y_with_noise = y + epsilon

x_train, x_test, y_train, y_test = train_test_split(x, y_with_noise, test_size=0.2, random_state=0)
lambda_values = np.arange(0, 5.1, 0.5)
train_errors = []
test_errors = []
non_zero_coeffs = []
l1_norms = []
learned_curves = []

for lmbda in lambda_values:
    lasso = Lasso(alpha=lmbda)
    lasso.fit(x_train.reshape(-1, 1), y_train)
    train_pred = lasso.predict(x_train.reshape(-1, 1))
    test_pred = lasso.predict(x_test.reshape(-1, 1))
    train_errors.append(mean_squared_error(y_train, train_pred))
    test_errors.append(mean_squared_error(y_test, test_pred))
    non_zero_coeffs.append(np.count_nonzero(lasso.coef_))
    l1_norms.append(np.linalg.norm(lasso.coef_, ord=1))
    learned_curves.append(lasso.predict(x.reshape(-1, 1)))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, train_errors, label='Training Error')
plt.plot(lambda_values, test_errors, label='Test Error')
plt.xlabel('Lambda')
plt.ylabel('Mean Squared Error')
plt.title('Lasso Regression: Training and Test Error')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, non_zero_coeffs)
plt.xlabel('Lambda')
plt.ylabel('Number of Non-zero Components')
plt.title('Lasso Regression: Number of Non-zero Components vs Lambda')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label='True Function')
plt.scatter(x_train, y_train, label='Training Data')
for i, lmbda in enumerate(lambda_values):
    plt.plot(x, learned_curves[i], label=f'Lambda={lmbda}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lasso Regression: True Function and Learned Curves')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, l1_norms)
plt.xlabel('Lambda')
plt.ylabel('L1 Norm of Coefficients')
plt.title('Lasso Regression: L1 Norm vs Lambda')
plt.grid(True)
plt.show()
```

```python
lambda_values = np.arange(0, 5.1, 0.5)
train_errors = []
test_errors = []
l2_norms = []
learned_curves = []

for lmbda in lambda_values:
    ridge = Ridge(alpha=lmbda)
    ridge.fit(x_train.reshape(-1, 1), y_train)
    train_pred = ridge.predict(x_train.reshape(-1, 1))
    test_pred = ridge.predict(x_test.reshape(-1, 1))
    train_errors.append(mean_squared_error(y_train, train_pred))
    test_errors.append(mean_squared_error(y_test, test_pred))
    l2_norms.append(np.linalg.norm(ridge.coef_))
    learned_curves.append(ridge.predict(x.reshape(-1, 1)))
```

```python
plt.figure(figsize=(10, 6))
plt.plot(lambda_values, train_errors, label='Training Error')
plt.plot(lambda_values, test_errors, label='Test Error')
plt.xlabel('Lambda')
plt.ylabel('Mean Squared Error')
plt.title('Ridge Regression: Training and Test Error')
plt.legend()
plt.grid(True)
plt.show()
```
